### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head(5)
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda1["Precio"].sum()
ingreso_total_tienda2 = tienda2["Precio"].sum()
ingreso_total_tienda3 = tienda3["Precio"].sum()
ingreso_total_tienda4 = tienda4["Precio"].sum()
```

```python
pd.DataFrame([ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4], columns=["Ingreso Total"], index=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"])
```

# 2. Ventas por categoría

```python
#tienda 1
ventas_por_categoria_tienda1 = tienda1['Categoría del Producto'].value_counts().reset_index()
ventas_por_categoria_tienda1.columns = ['Categoría del Producto', 'Cantidad de Productos Vendidos']
datos_categoria_tienda1 = [categoria for categoria in ventas_por_categoria_tienda1['Categoría del Producto']]
datos_ventas_tienda1 = [categoria for categoria in ventas_por_categoria_tienda1['Cantidad de Productos Vendidos']]


#tiienda 2
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts().reset_index()
ventas_por_categoria_tienda2.columns = ['Categoría del Producto', 'Cantidad de Productos Vendidos']
datos_categoria_tienda2 = [categoria for categoria in ventas_por_categoria_tienda2['Categoría del Producto']]
datos_ventas_tienda2 = [categoria for categoria in ventas_por_categoria_tienda2['Cantidad de Productos Vendidos']]

#tienda 3
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts().reset_index()
ventas_por_categoria_tienda3.columns = ['Categoría del Producto', 'Cantidad de Productos Vendidos']
datos_categoria_tienda3 = [categoria for categoria in ventas_por_categoria_tienda3['Categoría del Producto']]
datos_ventas_tienda3 = [categoria for categoria in ventas_por_categoria_tienda3['Cantidad de Productos Vendidos']]

#tienda 4
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts().reset_index()
ventas_por_categoria_tienda4.columns = ['Categoría del Producto', 'Cantidad de Productos Vendidos']
datos_categoria_tienda4 = [categoria for categoria in ventas_por_categoria_tienda4['Categoría del Producto']]
datos_ventas_tienda4 = [categoria for categoria in ventas_por_categoria_tienda4['Cantidad de Productos Vendidos']]

ventas_por_categoria_tienda1
```

```python
df = pd.DataFrame({'Categoría del Producto': datos_categoria_tienda1, 'Cantidad de Productos Vendidos': datos_ventas_tienda1})

df.set_index('Categoría del Producto')['Cantidad de Productos Vendidos'].plot(
    kind='pie', figsize=(8, 8), autopct='%1.1f%%', colormap='viridis')

plt.title("Distribución de Ventas por Categoría de Producto")
plt.ylabel('')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
tienda1_calificacion = tienda1["Calificación"].mean()
tienda2_calificacion = tienda2["Calificación"].mean()
tienda3_calificacion = tienda3["Calificación"].mean()
tienda4_calificacion = tienda4["Calificación"].mean()
```

```python
data = {'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        'Promedio Calificación': [tienda1_calificacion,tienda2_calificacion,tienda3_calificacion,tienda4_calificacion]}
df = pd.DataFrame(data)
df.plot(kind='bar', x='Tienda', y='Promedio Calificación',
        figsize=(8, 5), color='skyblue', legend=False)

plt.title('Promedio de Calificación por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Promedio de Calificación')
plt.ylim(0, 5)
plt.xticks(rotation=0)

plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_tienda1 = tienda1['Producto'].value_counts().reset_index()
productos_tienda1.columns = ['producto', 'Cantidad de Productos Vendidos']

productos_tienda2 = tienda2['Producto'].value_counts().reset_index()
productos_tienda2.columns = ['producto', 'Cantidad de Productos Vendidos']

productos_tienda3 = tienda3['Producto'].value_counts().reset_index()
productos_tienda3.columns = ['producto', 'Cantidad de Productos Vendidos']

productos_tienda4 = tienda4['Producto'].value_counts().reset_index()
productos_tienda4.columns = ['producto', 'Cantidad de Productos Vendidos']

producto_mas_vendido_tienda1 = productos_tienda1.head(1)
producto_mas_vendido_tienda2 = productos_tienda2.head(1)
producto_mas_vendido_tienda3 = productos_tienda3.head(1)
producto_mas_vendido_tienda4 = productos_tienda4.head(1)


producto_menos_vendido_tienda1 = productos_tienda1.tail(1)
producto_menos_vendido_tienda2 = productos_tienda2.tail(1)
producto_menos_vendido_tienda3 = productos_tienda3.tail(1)
producto_menos_vendido_tienda4 = productos_tienda4.tail(1)

```

```python
data = {'Tienda': [productos_tienda1["producto"].head(1),productos_tienda1["producto"].iloc[-1]], 'vendidos': [producto_mas_vendido_tienda1['Cantidad de Productos Vendidos'][0], producto_menos_vendido_tienda1['Cantidad de Productos Vendidos'].iloc[-1]]}
df = pd.DataFrame(data)


df.plot(kind='bar', x='Tienda', y='vendidos',
        figsize=(8, 5), color='skyblue', legend=False)


plt.title('Producto mas y menos vendido tienda 1')
plt.xlabel('Tienda1')
plt.ylabel('Cantidad de Productos Vendidos')
plt.ylim(0, max(df['vendidos']) + 5)
plt.xticks(rotation=0)


plt.show()
```

# 5. Envío promedio por tienda

```python
tienda1_costo_envio = tienda1["Costo de envío"].mean()
tienda2_costo_envio = tienda2["Costo de envío"].mean()
tienda3_costo_envio = tienda3["Costo de envío"].mean()
tienda4_costo_envio = tienda4["Costo de envío"].mean()
```

```python

data = {'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        'Costo de Envío Promedio': [tienda1_costo_envio, tienda2_costo_envio, tienda3_costo_envio, tienda4_costo_envio]}

df = pd.DataFrame(data)

df.plot(kind='barh', x='Tienda', y='Costo de Envío Promedio',
        figsize=(8, 5), color='red', legend=False)

plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envío')
plt.xticks(rotation=0)

plt.show()
```
